***Key Questions***

```python
## 1. Do men tend to prefer SUVs more compared to women?
plt.figure(figsize=(10,5))
sns.countplot(data=df,x="Gender",hue="Make")
plt.title("Gender Distribution for each car type")
plt.xlabel("Gender");
plt.ylabel("Count");
plt.legend(title="Make",loc="upper right", labels=["SUV","Sedan","Hatchback"])
plt.show()
```

```python
## 2.Likelihood of a Salaried person buying a Sedan?
plt.figure(figsize=(10,5))
plt.title("Salaried person buying Sedan");
sns.countplot(data=df,x="Profession",hue="Make");
```

```python
# Total number of Salaried individuals who brought Sedan
sedan_buyers = df.query("Profession == 'Salaried' and Make == 'Sedan'")
sedan_count=len(sedan_buyers)

# Total salaried individuals
total_salaried=df[df["Profession"] == "Salaried"].shape[0]

likelihood=(sedan_count/total_salaried) * 100

print("Likelihood of a Salaried person buying a Sedan",likelihood,'%')
```

```python
## 3.Sheldon Cooper claims that Salaried male is an easier target for SUV sale over Sedan sale?
plt.figure(figsize=(10,5))
sns.countplot(data=df,x="Make",hue="Gender",order=["SUV","Hatchback","Sedan"])
plt.title("Salaried car distribution for each car type");
plt.xlabel("Make")
plt.ylabel("Count")
plt.show()
```

```python
## 4.Amount spent on purchasing automobiles vary by gender

df.groupby("Gender")["Price"].describe()
```

```python
plt.figure(figsize=(10,5))
sns.catplot(data=df,x="Gender",y="Price",kind="box")
plt.title("Amount spent on purchasing automobiles vary by gender");
plt.xlabel("Gender");
plt.ylabel("Price");
plt.show()
```

```python
# Create dataframe for female customers
df_womens = df.loc[df['Gender'] == 'Female']
sns.boxplot(
    data=df_womens,
    x='Make',
    y='Price',
    hue='Make'
)

plt.xticks(rotation=45)
plt.title('Price Distribution by Make (Female Customers)')
plt.show()
```

```python
df_womens.groupby('Make')['Price'].describe()
```

```python
# Create dataframe for Male customers

df_males = df.loc[df['Gender'] == 'Male']

# Price distribution by Make (Male)
sns.boxplot(
    data=df_males,
    x='Make',
    y='Price'
)

plt.xticks(rotation=45)
plt.title('Price Distribution by Make (Male Customers)')
plt.show()
```
